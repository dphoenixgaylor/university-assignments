## Background ##

<b>Can meteorological variables such as wind, pressure, humidity, and temperature help predict air quality?</b><br />
<b>If not, is there at least a correlation between a county's average weather and its average air quality?</b> 

In this project, I extracted 77 data files containing the aforementioned meteorological variables and population variables by United States county. I then merged them together into a single dataframe, where I then performed linear regression techniques to try and capture the weather's effects on air quality.

## Extracting Data ##

From the EPA website, there are zip files containing information for one of my air quality or meteorological variables for one year. Because I have 5 variables for 15 years, I had to extract 75 files from the site using the code below. For obtaining population data, I simply downloaded the .csv file from the NBER website and for averaged coordinate data, I used a converter to convert a Wikipedia table to a .csv file.

<b>Step 1:</b> I imported several packages to help collect and organize the data I found.

```python
import requests, zipfile, io
import pandas as pd
```

<b>Step 2:</b> I created a function, <code>extract</code>, to download the .zip files on the EPA website and unpack them in a preselected directory.

```python
def extract(var, year):
    """Using a given variable and year, place corresponding extracted ZIP file in my Data folder."""
    
    r = requests.get('https://aqs.epa.gov/aqsweb/airdata/daily_' + var + '_' + str(year) + '.zip')
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall("/Users/Phoenix/Desktop/Gaylor_Final_STA141B/Data")

var_list = ['WIND', 'TEMP', 'PRESS', 'RH_DP', '88101']  # '88101' corresponds with the air quality data

[extract(var, year) for var in var_list for year in range(2000, 2015)]
print('Done!')
```

## Loading Data ##

Once I extracted the data on to my computer, I wanted to upload it into memory as six variables, combining 15 CSV files for the EPA variables and simply loading the other. I did so with the code below.

<b>Step 1:</b> I imported a couple more helpful packages, including a package that will help me filter warnings. Nobody likes warnings. :)

```python
import glob
import os
import warnings

warnings.filterwarnings('ignore')
```

<b>Step 2:</b> I created a function, <code>load</code>, that helped merge the similar .csv files into one variable in memory. I used this with all variables found from the EPA site.

```python
path = r'C:/Users/Phoenix/Desktop/Gaylor_Final_STA141B/Data'

def load(path, var):
    """Using my pre-selected path and a given variable, merge the CSV files of that variable and upload it into memory."""

    all_files = glob.glob(os.path.join(path, var + "*"))
    all_var = (pd.read_csv(f) for f in all_files)
    return pd.concat(all_var, ignore_index=True)

wind = load(path, 'daily_WIND')
temp = load(path, 'daily_TEMP')
press = load(path, 'daily_PRESS')
rhdp = load(path, 'daily_RH_DP')
aq = load(path, 'daily_88101')
```

<b>Step 3:</b> I uploaded the files I manually downloaded using Panda's <code>read_csv</code> function.

```python
population_raw = pd.read_csv("Data/county_population.csv", encoding = 'latin-1')
coordinates_raw = pd.read_csv("Data/county_coordinates.csv", encoding = "latin-1")
```

## Merging Data: *Meteorological & Air Quality Recordings* ##

One of the most difficults part of the project was merging the meteorological and air quality variables together. Because not every measuring center recorded every variable, there was a lot of missingness. Because the missingness of the data is random (one variable will not cause another to become missing), I believe it is OK to completely exclude them. However, successfully combining the data while matching each variable to its respective state, county, and date, was a challenge. The code to do it is below.

<b>Step 1:</b> Each variable dataframe had dozens of columns that were irrelevant to the problem at hand. For this reason, I created a new "main" dataframe for each variable. Each "main" only contained the <code>State Code</code>, <code>County Code</code>, <code>Date Local</code>, and <code>Arithmetic Mean</code> (the daily measurement). To avoid having 5 unrelated "Arithmetic Means", I renamed each to the variable's corresponding unit of measurement. I then ordered each dataframe by state, county, and date.

```python
wind_main = wind[['State Code', 'County Code', 'Date Local', 'Arithmetic Mean']]
wind_main.columns = ['State', 'County', 'Date', 'knots']
wind_main = wind_main.sort_values(by=['State', 'County', 'Date'])

temp_main = temp[['State Code', 'County Code', 'Date Local', 'Arithmetic Mean']]
temp_main.columns = ['State', 'County', 'Date', 'degF']
temp_main = temp_main.sort_values(by=['State', 'County', 'Date'])

press_main = press[['State Code', 'County Code', 'Date Local', 'Arithmetic Mean']]
press_main.columns = ['State', 'County', 'Date', 'mb']
press_main = press_main.sort_values(by=['State', 'County', 'Date'])

rhdp_main = rhdp[['State Code', 'County Code', 'Date Local', 'Arithmetic Mean']]
rhdp_main.columns = ['State', 'County', 'Date', 'rh']
rhdp_main = rhdp_main.sort_values(by=['State', 'County', 'Date'])

aq_main = aq[['State Code', 'County Code', 'Date Local', 'Arithmetic Mean']]
aq_main.columns = ['State', 'County', 'Date', 'mcg/m3']
aq_main = aq_main[~(aq_main['State'] == 'CC')]     # State Code CC does not have all variables and is not needed. By deleting it,
aq_main['State'] = pd.to_numeric(aq_main['State']) # and changing the column type, I can more easily merge the data sets.
aq_main = aq_main.sort_values(by=['State', 'County', 'Date'])
```

<b>Step 2:</b> At times, there were multiple entries indicating multiple measurements taken on a given date. Because I am primarily looking at daily measurements and extra entries are erroneous, I took the average of the multiple entries, so that there was only one entry per date per variable. I then merged all variables together into <code>all_avg</code>, deleting any data points that did not have entries for all 8 columns.

```python
wind_avg = pd.Series.to_frame(wind_main.groupby(['State', 'County', 'Date'])['knots'].mean())
temp_avg = pd.Series.to_frame(temp_main.groupby(['State', 'County', 'Date'])['degF'].mean())
press_avg = pd.Series.to_frame(press_main.groupby(['State', 'County', 'Date'])['mb'].mean())
rhdp_avg = pd.Series.to_frame(rhdp_main.groupby(['State', 'County', 'Date'])['rh'].mean())
aq_avg = pd.Series.to_frame(aq_main.groupby(['State', 'County', 'Date'])['mcg/m3'].mean())

all_avg = wind_avg.merge(temp_avg, on=['State', 'County', 'Date']).merge(press_avg, on=['State', 'County', 'Date']).merge(
    rhdp_avg, on=['State', 'County', 'Date']).merge(aq_avg, on=['State', 'County', 'Date'])
all_avg.reset_index(inplace=True) 
all_avg
```

## Visualizing Data: *Meteorological Variable Correlation* ##

After merging the data, I wanted to use a correlation heatmap to determine if any of the variables appeared to be correlated. If so, I could exclude one of them from my data, which would allow for a better prediction later on. In addition, if I can exclude a variable, I can attempt to merge the data again without one of the correlated variables and likely have more data points (since I am currently deleting data points that don't have all 5 variables).

I suspect that there may be some correlation between temperature, <code>degF</code> and relative humidity, <code>rh</code>, since when it is hotter, it is usually drier (at least in California!). I also suspect the air quality, <code>mcg/m3</code> to be slightly correlated with temperature and relative humidity, since from personal experience, when the temperature is high, air quality seems worse, and when it's raining (high relative humidity), air quality seems better. However, because I feel population will have a far larger impact on air quality than the weather, I don't expect the correlation to be strong.

<b>Step 1:</b> As usual, I utilized packages to help me get the job done.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

<b>Step 2:</b> Next, I subsetted the <code>all_avg</code> dataframe to only include the meteorological and air quality variables. I then made my correlation heatmap.

```python
all_avg_meteo = all_avg[['knots', 'degF', 'mb', 'rh', 'mcg/m3']]
corr = all_avg_meteo.corr()
corr_heatmap = sns.heatmap(corr, linewidths = .5, vmin = -0.3, vmax = 0.3, cmap="PiYG")
corr_heatmap.set_title('Meteorological Variable Correlation')
plt.show()
corr
```

### Insights from Correlation Heatmap ###

From the heatmap above, it seems that I was correct in my prediction that temperature and relative humidity may have a correlation, but the possible correlation is not significant enough to eliminate either variable. I did not consider wind, <code>knots</code>, and how it would be correlated with other variables, but there is a very slight correlation between wind, temperature, and even relative humidity. 

As I predicted, temperature and relative humidity do have a slight correlation with air quality, but again, wind also has a negative correlation. This makes sense, since when the wind is strong, pollutants may be pushed elsewhere.

Pressure, <code>mb</code>, seemed completely unrelated to all variables, including air quality. Because of the extreme lack of correlation between pressure and air quality (-0.023), I will exclude pressure from my data going forward and will then have many more relevant data points to work with.


Below, I construct a new dataframe, <code>rel_avg</code> (relevant averages) that exclude pressure. This new dataframe has an extra 91,000 data points to work with.

```python
rel_avg = wind_avg.merge(temp_avg, on=['State', 'County', 'Date']).merge(rhdp_avg, on=['State', 'County', 'Date']).merge(
    aq_avg, on=['State', 'County', 'Date'])
rel_avg.reset_index(inplace=True) 
rel_avg['Date'] = pd.to_datetime(rel_avg['Date'], format = '%Y-%m-%d')
```

## Merging Data: *Relevant Variables with Population & Coordinates* ##

The next step was to create a database of what state and county codes belonged with which latitude and longitude coordinates, along with the population by county from 2000 to 2015. Then, I combined this database with the relevant variables (rel_avg) to complete the merge of all data from all sources. <br /><br />
<i>Note: I was initially planning on using longitude and latitude for visuals using <code>basemap</code>. However, when I started to work on my visuals, I realized that it was not an appropriate tool to use. I have too many variables and too many data points to be displayed on one plot. I kept the work below to accurately show my whole process in completing this project and because later code relied on it.</i>

<b>Step 1:</b> I created <code>pop_main</code>, which has the population for each <code>State Code</code> and <code>County Code</code> from the years 2000 to 2015. <br /><br />
<b>Challenge:</b> The data was fragmented into two parts, data with a 1-3 digit <code>county_fips</code> (United States county identifier) and population for the years 2000-2009 and data with a 4-5 digit <code>county_fips</code> and the population for years 2010-2014. <br /><br />
Below, standardized <code>State</code> and <code>County</code> codes are created with the two fragmented sets merged together.

```python
pop_to2009 = population_raw[(population_raw['county_fips'] < 1000) & (population_raw['county_fips'] > 0)]
pop_to2009_main = pop_to2009[['state_fips', 'county_fips', 'pop2000', 'pop2001', 'pop2002', 'pop2003', 'pop2004', 'pop2005', 
                              'pop2006', 'pop2007', 'pop2008', 'pop2009']]
pop_to2009_main.columns = ['State', 'County', 'pop2000', 'pop2001', 'pop2002', 'pop2003', 'pop2004', 'pop2005', 'pop2006', 
               'pop2007', 'pop2008', 'pop2009']
pop_to2009_main = pop_to2009_main.sort_values(by=['State', 'County'])

pop_to2014 = population_raw[(population_raw['county_fips'] > 1000)]
pop_to2014 = pop_to2014[['county_fips', 'pop2010', 'pop2011', 'pop2012', 'pop2013', 'pop2014']]
pop_to2014['County'] = pop_to2014['county_fips'] % 1000
pop_to2014['State'] = (pop_to2014['county_fips'] / 1000).astype(int)
pop_to2014_main = pop_to2014.reindex(columns = ['State', 'County', 'pop2010', 'pop2011', 'pop2012', 'pop2013', 'pop2014'])
pop_to2014_main = pop_to2014_main.sort_values(by=['State', 'County'])

pop_main = pop_to2009_main.merge(pop_to2014_main,on=['State','County'])
```

<b>Step 2:</b> Next, I created <code>coord_main</code> which contains the coordinates of the center of each county. As mentioned earlier, this was to be used with <code>basemap</code>, which I later abandoned.

```python
coord_small = coordinates_raw[['FIPS', 'Latitude', 'Longitude']]

coord_small['County'] = coord_small['FIPS'] % 1000
coord_small['State'] = (coord_small['FIPS'] / 1000).astype(int)
coord_main = coord_small[['State', 'County', 'Latitude', 'Longitude']]
```

<b>Step 3:</b> I then combined <code>coord_main</code> with <code>pop_main</code> to create <code>county_db</code>, the complete coordinate and population database for each county.

```python
county_db = pop_main.merge(coord_main, on=['State', 'County'])
county_db = county_db.reindex(columns = ['State', 'County', 'Latitude', 'Longitude', 'pop2000', 'pop2001', 'pop2002', 
                                          'pop2003', 'pop2004', 'pop2005', 'pop2006', 'pop2007', 'pop2008', 'pop2009',
                                          'pop2010', 'pop2011', 'pop2012', 'pop2013', 'pop2014'])

county_db['Latitude'] = (county_db['Latitude'].map(lambda x: ''.join([i for i in x if i.isdigit()]))).astype(int).div(1000000)
county_db['Longitude'] = county_db['Longitude'].str[1:]
county_db['Longitude'] = (county_db['Longitude'].map(lambda x: ''.join([i for i in x if i.isdigit()]))).astype(int).div(1000000)
county_db
```

<b>Step 4:</b> I then created a function, <code>checkdb</code>, to add the population corresponding to the year of a data point.

```python
def checkdb(rel_avg_row):
    """Returns the population of a county at the time specified by the data point"""
    
    state = rel_avg_row['State']
    county = rel_avg_row['County']
    year = rel_avg_row['Date'].year
    
    return county_db[(county_db['State'] == state) & (county_db['County'] == county)]['pop' + str(year)].iloc[0]

rel_avg['Population'] = [checkdb(rel_avg.iloc[i]) for i in range(rel_avg.shape[0])]
```

## Statistical Analysis##

Now that all the data has been successfully merged, I was able to start my statistical analysis. I broke this up into two distinct analyses: an analysis of all data points in <code>rel_avg</code> and an analysis of the average of all data points in <code>rel_avg</code> by county.

```python
## packages to be used, in addition to the packages already imported

import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy.special import inv_boxcox
```

### Analysis of All Data Points ###

<b>Step 1</b>: I created <code>merged_data</code>, which will only use the meteorological and population variables of <code>rel_avg</code> and will get rid of impossible values and extreme outliers. Next, I created histograms to display the distribution of each variable.

```python
merged_data = rel_avg[['knots', 'degF', 'rh', 'Population', 'mcg/m3']]
merged_data = merged_data[(merged_data['knots'] >= 0) & (merged_data['rh'] >= 0) & (merged_data['rh'] <= 100) & 
                          (merged_data['mcg/m3'] >= 0) & (merged_data['mcg/m3'] <= 80)]
merged_data_hists = merged_data.hist(bins = 15, figsize = (16,10), layout = (2,3))
```

Because I deleted impossible values, the data looks correct from my brief check of the data set. With the exceptions of population and air quality, the distributions seem fairly symmetric. However, it makes sense that these variables are not symmetric, because air quality is typically not bad (low mcg/m3) and counties are typically fairly small (less than 100,000 people).

<b>Step 2</b>: I created a scatter matrix and a correlation heatmap to check for any correlation between the variables. A strong correlation would create multicolinearity problems when doing linear regression later.

```python
pd.scatter_matrix(merged_data, figsize = [12, 12], alpha = 0.1)
plt.show()
```

Based on the scatter matrix, none of the variables seem correlated.

```python
corr = merged_data.corr()
corr_heatmap = sns.heatmap(corr, linewidths = .5, vmin = -0.3, vmax = 0.3, cmap="PiYG")
corr_heatmap.set_title('Variable Correlation')
plt.show()
print(corr)
```

The correlation matrix shows that there are some very slight correlations between variables. While these correlations are extremely weak, I theorize:

* Temperature (<code>degF</code>) and relative humidity (<code>rh</code>) have a slight negative correlation because when it is raining (humidity is high), temperature tends to be lower.
* Population and temperature have a slight positive correlation because people do not like to live in very cold places (like Alaska).

<b>Step 3</b>: I created a linear model using all four potentially relevant variables. I then printed out the regression results summary.

```python
X = merged_data[['knots', 'degF', 'rh', 'Population']]
X = sm.add_constant(X)
Y = merged_data['mcg/m3']
 
model = sm.OLS(Y, X).fit()
print(model.summary())
```

The regression results show that the model only accounts for 5.2% of the variation in air quality. This is a bit disappointing, but not too surprising since weather and air quality have such high variations anyways. 

For warning 2, my research says that the condition number is likely large because of how population is not on the same scale as the other variables (population goes up to 1e+07, while other variables cap out just over 100). 

<b>Step 4:</b> To check for multicolinearity, I checked the VIFs, variance inflation factors, for each of my features.

```python
vif = pd.DataFrame()
vif["features"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif.round(2)
```

The VIF table above shows no multicolinearity, although the very low VIF for my constant indicates that it is not too great of a constant (VIF should be much, much higher).

<b>Step 5</b>: I found the different between the actual air quality and the predicted air quality to try and see how close the predictions typically were.

```python
merged_data['mcg/m3 predict'] = model.predict(X)
merged_data['Difference'] = merged_data['mcg/m3 predict'] - merged_data['mcg/m3']
merged_data['Difference'].describe()
```

From the above table, the predictions seem somewhat promising. The median difference between the real air quality and predicted air quality was 1.7 (really close!), with the middle 50% being between -2.47 and 4.52. This gave me confidence that despite the low r-squared, these variables may actually be a fair predictor of air quality.

<b>Step 6</b>: To test this theory, I then sorted the data by actual air quality, <code>mcg/m3</code>, and then plotted both the actual air quality and the predicted air quality.

```python
merged_data = merged_data.sort_values(by = 'mcg/m3')
merged_data = merged_data.reset_index(drop=True)
merged_data_plot = merged_data[['mcg/m3', 'mcg/m3 predict']].plot(title = 'Air Quality vs Predicted Air Quality', ylim = (0, 40))
```

Unfortunately, my hopes were lost when I saw this graph. According to the graph, the reason the predictor seemed to be close was because it constantly predicted within the most common air quality range (5 mcg/m3 and 20 mcg/m3). Since about 70% of the data fell in this range, the predictor was always decently close.

<b><u>Conclusion:</u> Based on my analysis of ALL data points, meteological variables CANNOT help predict air quality.</b>

### Analysis of Data Points Averaged by County ###

Because my analysis of all data points at once time showed that there is simply too much variation in air quality for it to be predicted with the weather, I move on to my next question. Is there at least a correlation between weather and air quality? In other words, if a county in the United States has certain average weather conditions, will they typically have a certain average air quality?

<b>Step 1:</b> To start this analysis, I grouped the <code>rel_avg</code> dataframe by state and county into a new dataframe, <code>rel_bySC</code> (relevant by state/county). This provided the average values for each county for all the variables.

```python
rel_bySC = rel_avg.groupby(['State', 'County']).mean()
```

<b>Step 2:</b> I then visualized the averaged variables' distribution with a series of histograms.

```python
rel_bySC_hists = rel_bySC.hist(bins = 15, figsize = (16,10), layout = (2,3))
```

The histograms above look very similar to when I took the histograms of all points. This makes sense, since when viewing a histogram, you can already visualize averages.

<b>Step 3:</b> I created a scatter matrix and a correlation heatmap to check for any correlation between the variables. Like before, a strong correlation would create multicolinearity problems when doing linear regression later.

```python
pd.scatter_matrix(rel_bySC[['knots', 'degF', 'rh', 'mcg/m3', 'Population']], figsize = [12, 12])
plt.show()
```

This time, when looking at the scatter matrix, it seems that some of the variables may have a slight positive correlation. It is somewhat hard to tell based on the scatter matrix, but the correlation matrix will likely shed some light on this. The correlations do not look significant enough to cause multicolinearity issues.

```python
corr = rel_bySC[['knots', 'degF', 'rh', 'mcg/m3', 'Population']].corr()
corr_heatmap = sns.heatmap(corr, linewidths = .5, vmin = -0.4, vmax = 0.4, cmap="PiYG")
corr_heatmap.set_title('Variable Correlation (when averaged by State & County)')
plt.show()
corr
```

As suspected, there is a fairly significant negative correlation between temperature, <code>degF</code>, and wind, <code>knots</code>. There is also a significant correlation between temperature and both air quality, <code>mcg/m3</code>, and population. Lastly, the most significant correlation is between population and air quality. <br />

<b><u>Conclusion:</u> Based on my analysis of averaged data points, there IS a correlation between average weather (notably temperature) and air quality.</b> However, correlation does not equate to causation. It is important to note that since temperature and population have a correlation, the variation in air quality may not be from a variation in temperature, but rather, population. <br />

### Continued Research: *Predicting a County's Average Air Quality with Average Temperature* ###

I am going to continue and try to see if it is possible to predict a county's average air quality from its average temperature. I will try predicting air quality with only temperature, only population, and then both temperature and population, to see if temperature has any noticeable benefit.

<b>Step 1:</b> I first created three different linear models for predicting air quality, <code>mcg/m3</code> with (a) temperature, (b) population, and (c) both temperature and population.

```python
Y = rel_bySC['mcg/m3']

X_temp = rel_bySC['degF']
X_temp = sm.add_constant(X_temp)
model_temp = sm.OLS(Y, X_temp).fit()

X_pop = rel_bySC['Population']
X_pop = sm.add_constant(X_pop)
model_pop = sm.OLS(Y, X_pop).fit()

X_both = rel_bySC[['degF', 'Population']]
X_both = sm.add_constant(X_both)
model_both = sm.OLS(Y, X_both).fit()
```

<b>Step 2:</b> I then created a new dataframe, <code>rel_byindex</code> which only included the three variables in question, and then added the predicted values based on the models above. I then dropped the index to have a smooth increasing plot for <code>mcg/m3</code>.

```python
rel_byindex = rel_bySC[['degF', 'Population', 'mcg/m3']].sort_values(by = 'mcg/m3')

rel_byindex['mcg/m3 predict (temp)'] = model_temp.predict(X_temp)
rel_byindex['mcg/m3 predict (pop)'] = model_pop.predict(X_pop)
rel_byindex['mcg/m3 predict (both)'] = model_both.predict(X_both)

rel_byindex = rel_byindex.reset_index(drop=True)
```

<b>Step 3:</b> I then plotted all the predictions alongside the actual <code>mcg/m3</code> values.

```python
rel_byindex_plot = rel_byindex[['mcg/m3', 'mcg/m3 predict (temp)', 'mcg/m3 predict (pop)', 'mcg/m3 predict (both)']].plot(
    title = 'Air Quality vs Predicted Air Quality')
```

From the graph above, it is clear that all three predictor models produce very similar results. This means that temperature is likely not the cause for a change in air quality and rather, it is the population. However, even population alone cannot accurately predict air quality.

* <b>HOWEVER</b>, there is some correlation between different meteological variables, especially temperature, and air quality, as seen by using models with the average variables for each county.
